```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('dataset.csv')
data.head()
```

```python
data.describe()
```

```python
import numpy as np
data[['Glucose_concentration', 'Blood_pressure', 'Insulin', 'BMI', 'Pedigree']
     ] = data[['Glucose_concentration', 'Blood_pressure', 'Insulin', 'BMI', 'Pedigree']].replace(0, np.NaN)
data.isna().sum()
data['Glucose_concentration'].fillna(data['Glucose_concentration'].mean(), inplace=True)
data['Blood_pressure'].fillna(data['Blood_pressure'].mean(), inplace=True)
data['Triceps'].fillna(data['Triceps'].median(), inplace=True)
data['Insulin'].fillna(data['Insulin'].median(), inplace=True)
data['BMI'].fillna(data['BMI'].median(), inplace=True)
data['Pedigree'].fillna(data['Pedigree'].median(), inplace=True)
import seaborn as sns
```

```python
plt.xticks(rotation=45)
def plot_box(data, cols, col_x='Class'):
    for col in cols:
        sns.set_style('whitegrid')
        sns.boxplot(data=data, x=col_x, y=col)
        plt.xlabel(col_x,)
        plt.ylabel(col)
        plt.show()
num_cols = list(data.columns)
plot_box(data, num_cols)
```

```python
X = data[num_cols]
y = data['Class']

from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X = scale.fit_transform(X)
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```
