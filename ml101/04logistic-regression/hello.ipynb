```python
import pandas as pd
f = pd.read_csv('dataset.csv')
print(f.shape)
print(f.head())
```

```python
f.isnull().sum()
f[f['Age'].isnull()]
```

```python
import numpy as np
age = f['Age'].values
age = np.reshape(age, (-1, 1))
age
```

```python
from sklearn.impute import SimpleImputer
import seaborn as sns
imp = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imp.fit(age)
f['Age'] = imp.transform(age)
f.corr()
```

```python
X = f[['Pclass', 'Fare', 'Sex_female', 'Sex_male', 'Age']]
X.head()

y = f['Survived']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
print(X_train.shape)
print(X_test.shape)
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'C': np.arange(1, 10, 0.5)}
lr_best = GridSearchCV(lr, parameters, cv=5)
lr_best.fit(X_train, y_train)
lr_best.best_estimator_
```

```python
from sklearn.metrics import classification_report
y_pred = lr_best.predict(X_test)
print('Acc ', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
